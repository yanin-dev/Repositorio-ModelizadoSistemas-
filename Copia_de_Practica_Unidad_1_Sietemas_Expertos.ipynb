```python
from experta import *

class EstadoObra(Fact):
    """Representa los datos sensoriales del sitio de construcción."""
    pass

class MotorSeguridad(KnowledgeEngine):

    # REGLA 1: RIESGO CRÍTICO (Prioridad Máxima)
    @Rule(OR(EstadoObra(viento=lambda v: v > 60),
             EstadoObra(grietas=True)),
          salience=100)
    def riesgo_critico(self):
        print("RESULTADO: RIESGO CRÍTICO.")
        print("ACCIÓN: PARO TOTAL DE OBRA. Evacuar zonas de altura y excavación.")
        self.declare(EstadoObra(estado_final='parado'))

    # REGLA 2: RIESGO MODERADO
    @Rule(AND(NOT(EstadoObra(estado_final='parado')),
              OR(EstadoObra(viento=lambda v: 40 <= v <= 60),
                 EstadoObra(humedad_extrema=True))))
    def riesgo_moderado(self):
        print("RESULTADO: RIESGO MODERADO.")
        print("ACCIÓN: PRECAUCIÓN. Suspender maniobras de izaje pesado.")

    # REGLA 3: BAJO RIESGO
    @Rule(AND(EstadoObra(viento=lambda v: v < 40),
              EstadoObra(grietas=False),
              EstadoObra(humedad_extrema=False)))
    def bajo_riesgo(self):
        print("RESULTADO: BAJO RIESGO.")
        print("ACCIÓN: OPERACIÓN NORMAL. Continuar según cronograma.")

# --- PRUEBA DEL SISTEMA ---
engine = MotorSeguridad()
engine.reset()

# Caso de prueba: Viento fuerte y detección de grietas
engine.declare(EstadoObra(viento=65, grietas=True, humedad_extrema=False))
engine.run()
```



El sistema utiliza un motor de inferencia basado en reglas mediante la librería experta.

El funcionamiento se basa en los siguientes componentes:

1. Hechos (Facts):
Se define una clase llamada EstadoObra que representa la información obtenida de los sensores, como:
- Velocidad del viento
- Presencia de grietas
- Nivel de humedad

2. Reglas:
Se definen reglas utilizando decoradores (@Rule) que representan condiciones específicas del sistema:
- Riesgo Crítico
- Riesgo Moderado
- Bajo Riesgo

Cada regla evalúa condiciones sobre los hechos ingresados.

3. Prioridad de reglas (salience):
Se utiliza el parámetro salience para definir la prioridad de ejecución de las reglas.
La regla de Riesgo Crítico tiene mayor prioridad para asegurarse de que se ejecute antes que cualquier otra condición.

4. Proceso de inferencia:
- Se ingresan los datos mediante declare()
- El motor evalúa todas las reglas disponibles
- Se ejecuta la regla que cumple las condiciones con mayor prioridad
- Se imprime el diagnóstico final

Este sistema permite tomar decisiones automáticas basadas en reglas, simulando el razonamiento lógico de un experto humano.

```python
from constraint import *

def planificar_maquinaria():
    problem = Problem()

    # 1. Definimos Variables (Máquinas) y Dominios (Zonas)
    maquinas = ["Grua_Torre", "Excavadora", "Hormigonera"]
    zonas = ["Zona_Estable", "Zona_Estrecha", "Zona_Comun"]

    problem.addVariables(maquinas, zonas)

    # 2. REGLAS DE ORO (Restricciones)

    # Restricción 1: Exclusividad (Una máquina por zona)
    problem.addConstraint(AllDifferentConstraint())

    # Restricción 2: Grúa Torre solo en Zona Estable (Base reforzada)
    problem.addConstraint(lambda gt: gt == "Zona_Estable", ["Grua_Torre"])

    # Restricción 3: Excavadora NO entra en Zona Estrecha
    problem.addConstraint(lambda ex: ex != "Zona_Estrecha", ["Excavadora"])

    # Restricción 4: Hormigonera NO puede estar con la Grúa (Congestión)
    # Al usar AllDifferent, ya están en zonas distintas, pero aquí
    # reforzamos que no pueden compartir recursos logísticos.
    problem.addConstraint(lambda ho, gt: ho != gt, ("Hormigonera", "Grua_Torre"))

    # 3. Resolución del Espacio del Problema
    soluciones = problem.getSolutions()

    print(f"Se encontraron {len(soluciones)} configuraciones seguras:")
    for i, sol in enumerate(soluciones, 1):
        print(f"Opción {i}: {sol}")

planificar_maquinaria()
```

En este problema, el uso de modelos basados en grafos y árboles resulta más adecuado que una estructura tradicional de if/else debido a la naturaleza del sistema y la complejidad de las decisiones.

Un enfoque basado en if/else trabaja de forma secuencial, evaluando condiciones una tras otra. Esto presenta varias limitaciones:
- Dificultad para escalar cuando aumentan las reglas.
- Mayor probabilidad de errores por el orden de evaluación.
- Código más rígido y difícil de mantener.
- No permite representar múltiples condiciones simultáneamente de forma clara.

En cambio, el uso de grafos y árboles, como ocurre en los sistemas expertos y problemas de restricciones:
- Permite representar el conocimiento mediante nodos (hechos) y conexiones (reglas).
- Evalúa múltiples reglas de forma más flexible, no estrictamente secuencial.
- Facilita la incorporación de nuevas reglas sin modificar toda la estructura.
- Permite priorizar decisiones mediante mecanismos como "salience".
- Modela mejor problemas reales donde intervienen múltiples variables al mismo tiempo.

En la Misión A (evaluación de riesgos), el sistema funciona como un árbol de decisión donde distintas condiciones llevan a distintos niveles de riesgo.

En la Misión B (planificación de maquinaria), el problema se representa como un grafo de restricciones, donde cada variable (máquina) debe cumplir condiciones sin entrar en conflicto con otras.

Por lo tanto, el modelo basado en grafos y árboles permite una solución más clara, escalable y mantenible que una simple lista de if/else.
