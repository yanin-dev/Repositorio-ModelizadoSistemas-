Una vez entrenado, utilizaremos el siguiente bloque para visualizar la Frontera de Decisión que tu modelo ha construido.

```python
# =========================
# PASO 1: LIBRERÍAS Y DATOS
# =========================

import tensorflow as tf  # Librería principal para redes neuronales
from tensorflow.keras import Sequential  # Modelo secuencial (una capa tras otra)
from tensorflow.keras.layers import Dense  # Capa densa (neurona)
import numpy as np  # Manejo de arrays numéricos
import matplotlib.pyplot as plt  # Para gráficos
from sklearn.datasets import make_blobs  # Para generar datos de prueba

# Generamos datos simulados (300 clientes, 2 características)
X, y = make_blobs(n_samples=300, centers=2, n_features=2, cluster_std=1.5)

# Graficamos los datos
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k')
plt.xlabel("Puntaje Crediticio")
plt.ylabel("Ratio de Ahorro")
plt.title("Distribución de Clientes")
plt.show()


# =========================
# PASO 2: MODELO (PERCEPTRÓN)
# =========================

# Creamos el modelo secuencial
model = Sequential([
    Dense(
        units=1,              # 1 neurona (porque es binario)
        activation='sigmoid', # Convierte salida entre 0 y 1 (probabilidad)
        input_shape=(2,)      # 2 variables de entrada
    )
])


# =========================
# PASO 3: COMPILACIÓN
# =========================

model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),  # Algoritmo de aprendizaje
    loss='binary_crossentropy',  # Error para clasificación binaria
    metrics=['accuracy']  # Métrica para ver qué tan bien aprende
)


# =========================
# PASO 4: ENTRENAMIENTO
# =========================

history = model.fit(
    X, y,           # Datos de entrada y salida
    epochs=50,      # Número de veces que aprende del dataset
    verbose=1       # Muestra el progreso en pantalla
)


# =========================
# PASO 5: VISUALIZAR FRONTERA
# =========================

def plot_frontier(X, y, model):
    # Definimos los límites del gráfico
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    # Creamos una grilla de puntos
    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, 0.1),
        np.arange(y_min, y_max, 0.1)
    )

    # Predecimos sobre cada punto de la grilla
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])

    # Reacomodamos los datos a la forma original
    Z = Z.reshape(xx.shape)

    # Dibujamos la frontera de decisión
    plt.contourf(xx, yy, Z, alpha=0.4, cmap='bwr')

    # Dibujamos los puntos reales
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='bwr')

    plt.title("Frontera de decisión del Perceptrón")
    plt.show()


# Llamamos a la función
plot_frontier(X, y, model)


# =========================
# EXTRA: VER PESOS Y BIAS
# =========================

pesos, bias = model.get_weights()  # Obtenemos pesos y bias

print("Pesos:", pesos)
print("Bias:", bias)
```

```python
# =========================
# PASO 1: LIBRERÍAS Y DATOS
# =========================

import tensorflow as tf  # Librería principal para redes neuronales
from tensorflow.keras import Sequential  # Modelo secuencial (una capa tras otra)
from tensorflow.keras.layers import Dense  # Capa densa (neurona)
import numpy as np  # Manejo de arrays numéricos
import matplotlib.pyplot as plt  # Para gráficos
from sklearn.datasets import make_blobs  # Para generar datos de prueba

# Generamos datos simulados (300 clientes, 2 características)
X, y = make_blobs(n_samples=300, centers=2, n_features=2, cluster_std=1.5)

# Graficamos los datos
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k')
plt.xlabel("Puntaje Crediticio")
plt.ylabel("Ratio de Ahorro")
plt.title("Distribución de Clientes")
plt.show()


# =========================
# PASO 2: MODELO (PERCEPTRÓN)
# =========================

# Creamos el modelo secuencial
from tensorflow.keras import regularizers

Dense(
    units=1,
    activation='sigmoid',
    input_shape=(2,),
    kernel_regularizer=regularizers.l2(0.01)
)


# =========================
# PASO 3: COMPILACIÓN
# =========================

model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),  # Algoritmo de aprendizaje
    loss='binary_crossentropy',  # Error para clasificación binaria
    metrics=['accuracy']  # Métrica para ver qué tan bien aprende
)


# =========================
# PASO 4: ENTRENAMIENTO
# =========================

history = model.fit(
    X, y,           # Datos de entrada y salida
    epochs=50,      # Número de veces que aprende del dataset
    verbose=1       # Muestra el progreso en pantalla
)


# =========================
# PASO 5: VISUALIZAR FRONTERA
# =========================

def plot_frontier(X, y, model):
    # Definimos los límites del gráfico
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    # Creamos una grilla de puntos
    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, 0.1),
        np.arange(y_min, y_max, 0.1)
    )

    # Predecimos sobre cada punto de la grilla
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])

    # Reacomodamos los datos a la forma original
    Z = Z.reshape(xx.shape)

    # Dibujamos la frontera de decisión
    plt.contourf(xx, yy, Z, alpha=0.4, cmap='bwr')

    # Dibujamos los puntos reales
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='bwr')

    plt.title("Frontera de decisión del Perceptrón")
    plt.show()


# Llamamos a la función
plot_frontier(X, y, model)


# =========================
# EXTRA: VER PESOS Y BIAS
# =========================

pesos, bias = model.get_weights()  # Obtenemos pesos y bias

print("Pesos:", pesos)
print("Bias:", bias)
```

"Se aplicó regularización L2 en la capa Dense para evitar el sobreajuste. Esto penaliza los pesos grandes, haciendo que el modelo sea más simple y generalice mejor. Como resultado, la frontera de decisión se vuelve más suave y los pesos disminuyen en magnitud."
